```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
#driver = webdriver.Chrome('chromedriver.exe')
driver = webdriver.Chrome(ChromeDriverManager().install()) #for mac
driver.get(url)
```

```python
src = driver.page_source
soup = BeautifulSoup(src, "html.parser")

data = []
table = soup.find('table', attrs={'class':'wikitable sortable jquery-tablesorter'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    try:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
    except:
        print("exception caught")
    

df = pd.DataFrame(data, columns = ['Rank',  'Name',  'Artist',  'Views',  'Upload Date', 'Note'])
df

```

```python
#Load
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import datetime
import os

from selenium.webdriver.chrome.service import Service #always import this or web driver gives truble
from selenium.webdriver.common.by import By #always import this or web driver gives truble
import time
from webdriver_manager.chrome import ChromeDriverManager # to manage the webdrivers for mac
```

```python
url = 'https://www.billboard.com'
#driver = webdriver.Chrome('chromedriver.exe')
driver = webdriver.Chrome(ChromeDriverManager().install()) #for mac
driver.get(url)

search_bar = driver.find_element(By.XPATH,'//*[@id="main-wrapper"]/header/div[2]/div/nav/ul/li[1]/a')
search_bar.click()
```

```python
src = driver.page_source
soup = BeautifulSoup(src, "html.parser")
song_name  = []
artist = []
last_week_rank = []
peak_rank = []
week_on_board = []

result = soup.find_all('div', class_='o-chart-results-list-row-container')
for res in result:
     try:
        song_name.append(res.find('h3').text.strip())
        artist.append(res.find('h3').find_next('span').text.strip())
        last_week_rank.append(res.find('h3').find_next('span').find_next('span').text.strip())
        peak_rank.append(res.find('h3').find_next('span').find_next('span').find_next('span').text.strip())
        week_on_board.append(res.find('h3').find_next('span').find_next('span').find_next('span').find_next('span').text.strip())
    except:
        print("exception caught")
    
    
song_name_100 = pd.Series(song_name)
artist_100 = pd.Series(artist)
last_week_rank_100 = pd.Series(last_week_rank)
peak_rank_100 = pd.Series(peak_rank)
week_on_board_100 = pd.Series(week_on_board)

top_100 = pd.concat([song_name_100, artist_100, last_week_rank_100, peak_rank_100, week_on_board_100], axis = 1)
top_100.rename(columns = {0: 'Song', 1: 'Artist', 2: 'Last week rank', 3: 'Peak Rank', 4: 'Week On Board'}, inplace = True)
top_100
```
